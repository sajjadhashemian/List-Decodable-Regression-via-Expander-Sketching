# Synthetic demo for ExpanderLDRRegressor
This notebook illustrates how to instantiate the placeholder estimator and run a basic synthetic experiment.

```python
import numpy as np
from expander_ldr.experiments import run_synthetic_experiment

def generator(n_samples, n_features):
    rng = np.random.default_rng(0)
    X = rng.standard_normal((n_samples, n_features))
    coef = rng.standard_normal(n_features)
    y = X @ coef + rng.normal(scale=0.1, size=n_samples)
    return X, y

result = run_synthetic_experiment(generator, n_samples=50, n_features=5)
result.coefficients
```
